```python
import cv2
import numpy as np
from PIL import ImageGrab
import pyautogui as au
import time

verdeB=np.array([50, 220, 20], np.uint8)
verdeA=np.array([70, 255, 255], np.uint8)
font = cv2.FONT_HERSHEY_SIMPLEX
f = []

diccionario = {'0' : (41,107), '1' : (107,174), '2':(107,107) , '3':(107,41) , '4':(174,174) , 
               '5': (174,107), 
               '6':(174,41) , '7': (240,174), '8':(240,107),'9':(240,41) , '10':(306,174) ,
               '11':(306,107) , '12':(306,41) ,'13':(373,174) , '14':(373, 107) , '15':(373,41) , '16':(439,174) ,
               '17': (439,107), '18': (439,41),'19':(505,174) , '20':(505,107) , '21':(505,41) , '22':(572,174) ,
               '23':(572, 107) , '24':(572,41) , '25':(638,174) , '26':(638,107) , '27':(638,41) , '28':(704,174),
               '29':(704,107) , '30':(704,41) , '31':(771,174) , '32':(771,107) , '33':(771,41) , '34':(837,174) , 
               '35':(837,107) , '36':(837,41)  }
girarX=236
girarY=561

rojoX = 851
rojoY = 931
negroX = 975
negroY = 931

def clear():
    au.moveTo(1297,992)
    au.click()
    
def rojoNegro():
    au.moveTo(rojoX,rojoY)
    au.click()
    time.sleep(0.1)
    au.moveTo(negroX,negroY)
    au.click()
    time.sleep(0.1)

def girar():
    au.moveTo(girarX,girarY)
    au.click()
def willy():   
    flag1 = False
    flag2 = False
    while flag2 == False:

        frame = np.array(ImageGrab.grab(bbox=(440,619,1318,833)))
        frame00 = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)


        frameHSV = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)


        mask = cv2.inRange(frameHSV,verdeB,verdeA)




        _,contornos,_ = cv2.findContours(mask, cv2.RETR_EXTERNAL,
          cv2.CHAIN_APPROX_SIMPLE)
        #cv2.drawContours(frame, contornos, -1, (255,0,0), 3)
        for c in contornos:
            area = cv2.contourArea(c)
            if area > 3000:
                M = cv2.moments(c)
                if (M["m00"]==0): M["m00"]=1
                x = int(M["m10"]/M["m00"])
                y = int(M['m01']/M['m00'])
                #print(x,y)
                cv2.circle(frame00, (x,y), 7, (0,255,0), -1)
                font = cv2.FONT_HERSHEY_SIMPLEX
                cv2.putText(frame00, '{},{}'.format(x,y),(x+10,y), font, 0.75,(0,255,0),1,cv2.LINE_AA)
                nuevoContorno = cv2.convexHull(c)
                cv2.drawContours(frame00, [nuevoContorno], 0, (255,0,0), 3)
                flag1 = True
            if flag1 == True:
                for name, xy in diccionario.items():
                    if xy == (x,y):
                            
                        print(name)
                        flag2 = True
                        f.append(int(name))
        #cv2.imshow('maskAzul',mask)
        #cv2.imshow('frame',frame)
        if cv2.waitKey(1) & 0xFF == ord('s'):
              break
time.sleep(3)
while True:
    rojoNegro()
    girar()
    time.sleep(1)
    clear()
    willy()  
    time.sleep(1)

                
cv2.destroyAllWindows()
```
